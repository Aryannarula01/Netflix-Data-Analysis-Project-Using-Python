## DATA ANALYTICS PROJECT 

```python
#importing libraries

import pandas as pd
import matplotlib.pyplot as plt
```

## Data Cleaning

```python
df=pd.read_csv("netflix_titles.csv")
#checking for null values
df.isnull().sum()
```

```python
#Deleting or droping null values
df.dropna(subset=["director","cast","country","date_added","rating","duration"],inplace=True)

#Checking again for null values
df.isnull().sum()
```

```python
#Overview of the dataset
df.head()
```

## Now, EDA (Exploratroy Data Analysis)

```python
#TV Shows v/s Movies

#Bar graph
type_count = df['type'].value_counts()
plt.bar(type_count.index,type_count.values,color=["skyblue","orange"])
plt.xlabel("Types")
plt.ylabel("Counts")
plt.tight_layout()
plt.title("TV Shows v/s Movies")
plt.show()
```

```python
#Chart 2  (Rating Counts)

rating_counts = df['rating'].value_counts().head(5)

#pie chart
plt.pie(rating_counts,labels=rating_counts.index,autopct='%1.1f%%')
plt.title("Top 5 Categories of Movies Rating")
plt.tight_layout()
plt.show()
```

```python
#chart 3 (Movies Duration Chart)

movies_duration = df[df['type'] == 'Movie'].copy()
movies_duration['duration int'] = movies_duration['duration'].str.replace(" min", "").astype(int)

#Histogram 
plt.hist(movies_duration['duration int'], bins=20, color='purple', edgecolor='black')
plt.title("Movies Duration Chart")
plt.xlabel("Minutes")
plt.ylabel("Number of Movies")
plt.tight_layout()
plt.show()

```

```python
#chart 4 (Release Year v/s Number of shows)

shows_count = df['release_year'].value_counts().sort_index()

plt.figure(figsize=(10,6))

#Scatter Chart
plt.scatter(shows_count.index,shows_count.values,color='blue',marker='o')
plt.grid(linestyle ="--",linewidth=1,color='gray')
plt.title("Release Year v/s Number of shows")
plt.xlabel("Release Year")
plt.ylabel("Number of Shows")
plt.tight_layout()
plt.show()
```

```python
# Top 10 country with number of shows count

top_country = df['country'].value_counts().head(10)
plt.figure(figsize=(10,6))

#Horizontal Bar Graph
plt.barh(top_country.index,top_country.values,color='red')
plt.xlabel("Number of Shows")
plt.ylabel("Countries")
plt.title("Number of shows by Country")
plt.tight_layout()
plt.show()
```
